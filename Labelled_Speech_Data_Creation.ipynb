###DOWNLOADING AN AUDIO FROM WEB TO TRANSCRIBE

```python
AUDIO_LINK = 'https://www.youtube.com/watch?v=26NHPSRIVrs'
start = 0
end = 437
```

```python
!python3 -m pip install -U "yt-dlp[default]"
```

```python
from yt_dlp import YoutubeDL

ffmpeg_args = {
  # - Don't forget the _i after "ffmpeg"; this puts the arguments before ffmpeg's `-i` argument,
  #   thus short-circuiting the download itself. Fail to do that,
  #   and you might as well skip ffmpeg for the download and trim in post-processing.
  # - Note that the arguments are pre-parsed into a list, like you'd pass to `subprocess.run`.
  "ffmpeg_i": ["-ss", str(start), "-to", str(end)]
}

opts = {
  "external_downloader": "ffmpeg",
  "external_downloader_args": ffmpeg_args,
  'extract_audio': True,
  'format': 'bestaudio',
  'outtmpl': 'audio.mp3'
}

def download_audio(link):
  with YoutubeDL(opts) as video:
    info_dict = video.extract_info(link, download = True)
    video_title = info_dict['title']
    print(video_title)
    video.download(link)
    print("Successfully Downloaded - see local folder on Google Colab")

download_audio(AUDIO_LINK)
```

###SEGMENTATION

```python
!pip install pydub
```

```python
#Generate random slices between 3 and 10 seconds
a = 4000
b = 7000
export_loc = '/content/sliced_audio/'
```

```python
from pydub import AudioSegment
import os
```

```python
audio = AudioSegment.from_file(filename)
```

RANDOM LENGTH SEGMENTATION (slices between 5 and 10 seconds)

```python
import random

total_duration = len(audio) #in millisecond
audio_slices = []
end_duration = 0

#First slice
start = 0
stop = 0
slice_duration = random.randint(a, b)
stop =  stop + slice_duration

while stop <= total_duration:
  audio_slices.append(audio[start:stop])
  slice_duration = random.randint(a, b)
  start = stop
  stop = stop + slice_duration

audio_slices.append(audio[start:total_duration])

counter = 0 #counter for no. of audio slices
export_format = "mp3"
try:
  os.makedirs(export_loc)
except FileExistsError:
  print('Folder already exists. Skipping folder creation.')


for audio_slice in audio_slices:
  counter += 1;
  audio_slice.export(f"{export_loc}/audio_{counter}.mp3", format = export_format)
```

###TRANSCRIBE BASED ON SEGMENTS FROM WHISPER

```python
file = 'About Love Loknath Koirala.txt'
f_path = '/content/'

from pathlib import Path
import json

with open(f_path + file, 'r') as f:
    f_str = f.read()
    f_str = f_str.replace("'", '"')
    f_str = f_str.replace('"s ', "'s ")
    f_str = f_str.replace('n"t ', "n't ")
    trans_result = json.loads(f_str)
```

```python
import torchaudio

wav, sample_rate = torchaudio.load('/content/audio.mp3')
```

```python
print(sample_rate)
print(len(wav))
```

```python
import torch
def convert_to_mono(audio_tensor: torch.Tensor) -> torch.Tensor:
        if audio_tensor.shape[0] == 1:
            return audio_tensor
        return torch.mean(audio_tensor, dim=0, keepdim=True)
```

```python
mono_wav = convert_to_mono(wav)
```

```python
print(mono_wav.shape[1])
print(mono_wav.shape[1]/sample_rate)
```

```python
out_dir = '/content/clips'
transcr_f = 'transcription.txt'
audio_f = 'L_Koirala_Chekhov_audio'
with open(f"{out_dir}/{transcr_f}", mode = 'w', encoding = 'utf-8') as tr_f:
    for segment in trans_result['segments']:
        sample_start = int(segment["start"]*sample_rate)
        sample_end = int(segment["end"]*sample_rate)
        # print(f"audio_{segment["id"]}\t{sample_start}\t{sample_end}\t{segment['text']}")
        tr_f.write(f"{audio_f}_{segment['id']}\t{segment['text']}\n")
        print(f"Saving {out_dir}/{audio_f}_{segment["id"]}")
        torchaudio.save(uri= f"{out_dir}/{audio_f}_{segment["id"]}.mp3", src = mono_wav[:,sample_start:sample_end], sample_rate = sample_rate )
```
